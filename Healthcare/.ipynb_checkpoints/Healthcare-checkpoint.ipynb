```python
###For accounts discharged during 2019, which were the most common primary diagnoses and Secondary diagnoses?###

# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files into pandas DataFrames
diagnosis_df = pd.read_csv('./Transactional Data/diagnosis.csv')
accounts_df = pd.read_csv('./Transactional Data/accounts.csv')
lookup_table_diagnosis_df = pd.read_csv('./Reference Data/lookup_table_diagnosis.csv')

##Primary diagnosis start##
# Filter accounts discharged during 2019
accounts_2019 = accounts_df[accounts_df['Discharge Year'] == 2019]

# Join diagnosis and accounts DataFrames
merged_df = pd.merge(diagnosis_df, accounts_2019, on='Account ID', how='inner')

# Filter primary diagnoses and group by diagnosis code to get frequency
primary_diagnoses = merged_df[merged_df['Diagnosis Sequence'] == 1]
diagnosis_frequency = primary_diagnoses['Diagnosis Code'].value_counts().reset_index()

# Rename columns
diagnosis_frequency.columns = ['Diagnosis Code', 'Frequency']

# Merge with lookup table to get diagnosis descriptions
result_df = pd.merge(diagnosis_frequency.head(10), lookup_table_diagnosis_df, on='Diagnosis Code', how='left')

# Select relevant columns and format the diagnosis code and description
result_df['Diagnosis'] = result_df['Diagnosis Code'] + ' (' + result_df['Diagnosis Description'] + ')'
result = result_df[['Diagnosis', 'Dx Clinical Classification', 'Frequency']]
#print(result)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(result['Diagnosis'], result['Frequency'], color='skyblue')
plt.xlabel('Diagnosis')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Primary Diagnoses')
plt.xticks(rotation=35, ha='right')
plt.tight_layout()
plt.show()

##Primary diagnosis end##

##Secondary diagnosis start##
# Filter secondary diagnoses and group by diagnosis code to get frequency
secondary_diagnoses = merged_df[merged_df['Diagnosis Sequence'] > 1]
diagnosis_frequency_secondary = secondary_diagnoses['Diagnosis Code'].value_counts().reset_index()

# Rename columns
diagnosis_frequency_secondary.columns = ['Diagnosis Code', 'Frequency']

# Merge with lookup table to get diagnosis descriptions
result_df_secondary = pd.merge(diagnosis_frequency_secondary.head(10), lookup_table_diagnosis_df, on='Diagnosis Code', how='left')

# Select relevant columns and format the diagnosis code and description
result_df_secondary['Diagnosis'] = result_df_secondary['Diagnosis Code'] + ' (' + result_df_secondary['Diagnosis Description'] + ')'
result_secondary = result_df_secondary[['Diagnosis', 'Dx Clinical Classification', 'Frequency']]

# Plotting the bar chart for secondary diagnoses
plt.figure(figsize=(10, 6))
plt.bar(result_secondary['Diagnosis'], result_secondary['Frequency'], color='lightgreen')
plt.xlabel('Diagnosis')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Secondary Diagnoses')
plt.xticks(rotation=35, ha='right')
plt.tight_layout()
plt.show()

##Secondary diagnosis end##
```

```python
###Which diagnosis related groups (MS-DRGs) have the highest average charges per patient account?###

# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files into pandas DataFrames
accounts_df = pd.read_csv('./Transactional Data/accounts.csv')
charges_df = pd.read_csv('./Transactional Data/charges.csv')
lookup_table_MS_DRG_df = pd.read_csv('./Reference Data/lookup_table_MS DRG.csv')

# Merge charges with accounts to get MS-DRG information
merged_df = pd.merge(charges_df, accounts_df, on='Account ID', how='inner')

# Merge with lookup table to get MS-DRG descriptions
merged_df = pd.merge(merged_df, lookup_table_MS_DRG_df, on='MS DRG Code', how='left')

# Group by account ID, MS-DRG code, and MS-DRG description and then calculate average total charges
result_df = merged_df.groupby(['Account ID', 'MS DRG Code', 'MS DRG Description']).agg(avg_total_charges=('Total Charges', 'mean')).reset_index()

# Sort the result by average total charges in descending order
result_df = result_df.sort_values(by='avg_total_charges', ascending=False)

# Print the top records to verify
#print(result_df.head(10))

# Limit the number of MS-DRG groups shown on the chart
top_n = 10

# Sort the DataFrame by average total charges in descending order and select the top 10 rows
top_n_df = result_df.head(top_n)

# Plotting the bar chart for the top 10 MS-DRG groups
plt.figure(figsize=(15,15))
plt.bar(top_n_df['MS DRG Description'], top_n_df['avg_total_charges'], color='skyblue')
plt.xlabel('MS DRG Description')
plt.ylabel('Average Total Charges')
plt.title(f'Top {top_n} MS DRG Groups by Average Total Charges')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
###Which Attending Provider Service had the highest volume of accounts (encounters) during December 2018?###

import pandas as pd

# Read the CSV files into pandas DataFrames
accounts_df = pd.read_csv('./Transactional Data/accounts.csv')
charges_df = pd.read_csv('./Transactional Data/charges.csv')
lookup_table_provider_df = pd.read_csv('./Reference Data/lookup_table_provider.csv')


# Filter accounts for December 2018
filtered_accounts = accounts_df[accounts_df['Admit Date'].str.startswith('2018-12')]

# Merge with lookup table to get Provider_Service
merged_df = pd.merge(filtered_accounts, lookup_table_provider_df, left_on='Attending Provider ID', right_on='Provider ID', how='inner')

# Group by Attending_Provider_ID and calculate count of account_id and maximum Provider_Service
result_df = merged_df.groupby('Attending Provider ID').agg(account_count=('Account ID', 'count'), Provider_Service=('Provider Service', 'max')).reset_index()

# Sort by account_count in descending order
result_df = result_df.sort_values(by='account_count', ascending=False)

# Select the first row
result_df = result_df.head(1)

# Select the columns you want to display
result_df = result_df[['account_count', 'Provider_Service']]

# Display the result without index
print(result_df.to_string(index=False))

```

```python
###What is the “average length of stay” (number of hospital days between admit date and discharge date, 
###but not including discharge date) for each of the 4 facilities for each of the years for which account 
###information are provided (2018 & 2019).###

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files into pandas DataFrames
accounts_df = pd.read_csv('./Transactional Data/accounts.csv')
lookup_table_facility_df = pd.read_csv('./Reference Data/lookup_table_facility.csv')

# Convert admit_date to datetime format
accounts_df['Admit Date'] = pd.to_datetime(accounts_df['Admit Date'])

# Filter accounts for years 2018 and 2019
filtered_accounts = accounts_df[accounts_df['Admit Date'].dt.year.isin([2018, 2019])]

# Merge with lookup table to get Facility_Name
merged_df = pd.merge(filtered_accounts, lookup_table_facility_df, on='Facility Code', how='inner')

# Calculate average length of stay rounded to 2 decimal places
result_df = merged_df.groupby(['Facility Name', merged_df['Admit Date'].dt.year.rename('year')]).apply(lambda x: round(x['LOS'].mean() - 1 if x['Discharge Date'].nunique() > 1 else x['LOS'].mean(), 2)).reset_index(name='avg length of stay')

# Pivot the DataFrame for better visualization
pivot_df = result_df.pivot(index='Facility Name', columns='year', values='avg length of stay')

# Plotting the bar chart
pivot_df.plot(kind='bar', figsize=(10, 6))
plt.title('Average Length of Stay by Facility (2018-2019)')
plt.xlabel('Facility Name')
plt.ylabel('Average Length of Stay (days)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
###Which MS-DRGs Service Lines had the largest difference between the observed length of stay and the national average provided in the MS-DRG reference file? 
###Use the "Arithmetic Mean LOS" field as the national average.###

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files into pandas DataFrames
accounts_df = pd.read_csv('./Transactional Data/accounts.csv')
lookup_table_MS_DRG_Service_Line_df = pd.read_csv('./Reference Data/lookup_table_MS DRG Service Line.csv')
lookup_table_MS_DRG_df = pd.read_csv('./Reference Data/lookup_table_MS DRG.csv')

# Merge accounts with MS-DRG service line lookup table
merged_df = pd.merge(accounts_df, lookup_table_MS_DRG_Service_Line_df, on='MS DRG Code', how='inner')

# Merge with MS-DRG reference table to get national average length of stay
merged_df = pd.merge(merged_df, lookup_table_MS_DRG_df[['MS DRG Code', 'Arithmetic Mean LOS']], on='MS DRG Code', how='inner')

# Group by service line description and national average length of stay, and calculate observed average length of stay
result_df = merged_df.groupby(['Service Line Description', 'Arithmetic Mean LOS']).agg({
    'LOS': 'mean'
}).reset_index()

# Calculate the absolute difference between observed and national average length of stay
result_df['Difference'] = result_df['LOS'] - result_df['Arithmetic Mean LOS']
result_df['Difference'] = result_df['Difference'].abs()

# Sort by difference in descending order and select top 10
result_df = result_df.sort_values(by='Difference', ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(result_df['Service Line Description'], result_df['Difference'], color='skyblue')
plt.xlabel('Service Line Description')
plt.ylabel('Absolute Difference (Days)')
plt.title('Top 10 Service Lines with Highest Absolute Difference in Length of Stay')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
###view of the monthly patient admissions over the years###

import pandas as pd
import matplotlib.pyplot as plt
import calendar

# Read the data into a pandas DataFrame
accounts_df = pd.read_csv('./Transactional Data/accounts.csv')

# Convert 'admit_date' to datetime format
accounts_df['Admit Date'] = pd.to_datetime(accounts_df['Admit Date'])

# Filter data for the years 2018 and 2019
filtered_accounts_df = accounts_df[accounts_df['Admit Date'].dt.year.isin([2018, 2019])].copy()

# Extract month and year from 'Admit Date' using .loc
filtered_accounts_df.loc[:, 'month'] = filtered_accounts_df['Admit Date'].dt.month_name()
filtered_accounts_df.loc[:, 'year'] = filtered_accounts_df['Admit Date'].dt.year

# Group by month and year, and count the number of patients
result_df = filtered_accounts_df.groupby(['month', 'year']).size().reset_index(name='patient_count')

# Sort the result by year and month
result_df['month'] = pd.Categorical(result_df['month'], categories=list(calendar.month_name)[1:], ordered=True)
result_df = result_df.sort_values(by=['year', 'month'])
#print(result_df)

# Pivot the DataFrame to have months as columns and years as index
pivot_df = result_df.pivot(index='month', columns='year', values='patient_count')

# Plotting the line chart
pivot_df.plot(kind='line', marker='o', figsize=(10, 4), color=['blue', 'green'])
plt.title('Patient Admissions by Month (2018-2019)')
plt.xlabel('Month')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45) 
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```
